```python
from urllib.request import urlopen
```

```python
from bs4 import BeautifulSoup as bs
```

```python
html = urlopen("https://movie.naver.com/movie/bi/mi/point.nhn?code=172780#")
```

```python
bsObject = bs(html, "html.parser")
```

```python
print(bsObject)
```

```python
###네이버 평점 URL 페이지 가져와서 html 오픈하고 웹문서 전체 출력
```

```python
all_divs = bsObject.find_all("iframe", {'class':'ifr'})
```

```python
print(all_divs)
```
