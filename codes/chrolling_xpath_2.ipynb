```python
from selenium import webdriver as wd
import time
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
```

```python
chromedriver_path = 'D:/venv/WORKSPACE\Chrolling/libraries/chromedriver.exe'
all_reple = set()
neg_arr = set()
pos_arr = set()
```

```python
def chrollingReple(driver, all_reple, neg_arr, pos_arr):
    firstSen = ""
    for i in range(1, 11):
        html = "/html/body/div[@class='ifr_area basic_ifr']/div[@class='input_netizen ']/div[@class='score_result']/ul/li["+ str(i) +"]/div[@class='score_reple']/p/span[@id='_filtered_ment_"+str(i-1)+"']"
        score_html = "/html/body/div[@class='ifr_area basic_ifr']/div[@class='input_netizen ']/div[@class='score_result']/ul/li["+ str(i) +"]/div[@class='star_score']/em"
        id_reple_html = "/html/body/div[@class='ifr_area basic_ifr']/div[@class='input_netizen ']/div[@class='score_result']/ul/li["+ str(i) +"]/div[@class='score_reple']/dl/dt/em/a/span"
        spoil_html = "/html/body/div[@class='ifr_area basic_ifr']/div[@class='input_netizen ']/div[@class='score_result']/ul/li["+ str(i) +"]/div[@class='score_reple']/p/span[@id='_text_spo_'"+str(i-1)+"']"
        reple = driver.find_element_by_xpath(html)
        score = driver.find_element_by_xpath(score_html)
        id_reple = driver.find_element_by_xpath(id_reple_html)
        reple_arr = set()
        if i == 1:
            firstSen = reple.text
        if reple.text == "":
            continue 
        if score.text == "10" or score.text == "9":
            try:
                spoil = driver.find_element_by_xpath(spoil_html)
            except NoSuchElementException:
                reple_arr.add(id_reple.text)
                reple_arr.add(reple.text)
                reple_arr.add(1)
                pos_arr.add(reple.text)
                print("\n긍정 텍스트 개수 : " + str(len(pos_arr)))
                print(reple.text)
        elif score.text == "0" or score.text == "1" or score.text == "2":
            try:
                spoil = driver.find_element_by_xpath(spoil_html)
            except NoSuchElementException:
                reple_arr.add(id_reple.text)
                reple_arr.add(reple.text)
                reple_arr.add(0)
                neg_arr.add(reple.text)
                print("\n부정 텍스트 개수 : " + str(len(neg_arr)))
                print(reple.text)
        else:
            continue
        all_reple.add(tuple(reple_arr))
        print("모은 총 텍스트 개수 : " + str(len(all_reple)))
    return firstSen
```

```python
def chrollingPage(index):
    ChrollingCount = 0
    while(True):
        ChrollingCount += 1
        html_url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code="+ str(index) +"&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=" + str(ChrollingCount)
        driver.get(html_url)
        try:
            if ChrollingCount % 2 == 1:
                firstSent = chrollingReple(driver, all_reple, neg_arr, pos_arr)
            elif ChrollingCount %2 == 0:
                secondSent = chrollingReple(driver, all_reple, neg_arr, pos_arr)
                if firstSent == secondSent:
                    index += 1
                    chrollingPage(index)
                    break
        except NoSuchElementException:
            index += 1
            chrollingPage(index)
            break
```

```python
driver = wd.Chrome(chromedriver_path)
chrollingPage(153516)
```
